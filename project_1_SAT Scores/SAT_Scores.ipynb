##### 12. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python

def my_avg(x):
    return sum(x)*1.0/len(x)


def my_std(x):
    avg = my_avg(x)
    return (sum([(elm-avg)**2 for elm in x])/len(x))**0.5
```

```python
    
print 'Rate:', my_std(data_dict['rate'])
print 'Math:', my_std(data_dict['math'])
print 'Verbal:', my_std(data_dict['verbal'])
```
